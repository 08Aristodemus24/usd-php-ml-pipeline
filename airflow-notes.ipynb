# There are essentially two ways to run airflow the first but ime the inefficient way is by using **ubuntu distribution** under the Windows subsystem for Linux if you are using windows. And the other more efficient way is by using **docker**. FF. covers the former way

* Luckily, Windows provides a convenient utility for using Linux along side Windows. This utility is called WSL (Windows Subsystem for Linux). Its recent version is WSL2
* Windows Subsystem for Linux provides a compatibility layer that lets you run Linux binary executables natively on Windows
* First, enable the windows subsystem for Linux option in settings. Go to Start. Search for "Turn Windows features on or off." Check the option Windows Subsystem for Linux.
* run `wsl --install` which may output the following or just right away by default install ubuntu

# Using docker


1. download docker desktop
2. sign up to docker with google or github account
3. enable wsl (windows subsystem for linux by searching turn windows features on/off and checking Windows Subsystem for Linux)
4. restart
5. change version of wsl from 1 to 2 if this is the case by `wsl --set-default-version 2`. You can look at the distribution being used by `wsl --status`. 
6. open docker desktop
7. go to your main working directory of your DE project and run `curl -LfO "https://airflow.apache.org/docs/apache-airflow/2.10.4/docker-compose.yaml"` to create the `.yaml` file and be able to run airflow using docker
8. creawte logs, plugins, config, and dags directory by running `mkdir -p ./logs ./plugins ./config` if in linux or `mkdir logs plugins config dags` if in windows
9. Initialize the database On all operating systems, you need to run database migrations and create the first user account. To do this, run `docker-compose up airflow-init`. Once successful terminal will show the ff.
```
airflow-init_1       | Upgrades done
airflow-init_1       | Admin user airflow created
airflow-init_1       | 2.10.4
start_airflow-init_1 exited with code 0
```
10. run `docker compose up` 
11. go to address http://0.0.0.0:8080 or http://localhost:8080
12. login with the credentials, username as `airflow`, password as `airflow`
13. now we stop the airflow webserver, scheduler, and workers by typing `ctrl + c` and then runnig `docker compose down --volumes --remove-orphans` to clean it up and restart from scratch removing all the existing dags
14. then we go to the docker-compose.yaml file and set the value of the `AIRFLOW__CORE__LOAD_EXAMPLES` key to `false`
